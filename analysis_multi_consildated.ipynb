```python
import pandas as pd

multi_baseline_df = pd.read_csv('analysis_implicit_hate_detection_multi_class_seed_model.csv', usecols = ['post','gt_labels_baseline', 'pred_labels_baseline'], names = ['index', 'post','gt_labels_baseline', 'pred_labels_baseline'], header=0,)
multi_bertweet_df = pd.read_csv('analysis_implicit_hate_detection_multi_class_bertweet.csv', usecols = ['post','gt_labels_bertweet', 'pred_labels_bertweet'], names = ['index', 'post','gt_labels_bertweet', 'pred_labels_bertweet'], header=0,)
multi_bertweet_hsol_df = pd.read_csv('analysis_implicit_hate_detection_multi_class_bertweet_and_hsol.csv', usecols = ['post','gt_labels_bertweet_hsol', 'pred_labels_bertweet_and_hsol'], names = ['index', 'post','gt_labels_bertweet_hsol', 'pred_labels_bertweet_and_hsol'], header=0,)
multi_berttweet_sarc_df = pd.read_csv('analysis_implicit_hate_detection_multi_class_bertweet_and_sarc.csv', usecols = ['post','gt_labels_bertweet_sarc', 'pred_labels_bertweet_sarc'], names = ['index', 'post','gt_labels_bertweet_sarc', 'pred_labels_bertweet_sarc'], header=0,)
```

```python
test_results = pd.merge(multi_baseline_df, multi_bertweet_df, how='inner', on=['post'])
test_results
```

```python
test_results_final = pd.merge(test_results, multi_bertweet_hsol_df, how='inner', on=['post'])
test_results_final
```

```python
baseline_false_predictions = test_results_final[test_results_final['gt_labels_baseline'] != test_results_final['pred_labels_baseline']]
baseline_false_predictions
```

```python
bertweet_correct_predictions = baseline_false_predictions[baseline_false_predictions['gt_labels_bertweet'] == baseline_false_predictions['pred_labels_bertweet']]
bertweet_correct_predictions
```

```python
bertweet_correct_predictions.to_csv('bertweet_correct_predictions.csv')
```

```python
bertweet_hsol_correct_predictions = baseline_false_predictions[baseline_false_predictions['gt_labels_bertweet_hsol'] == baseline_false_predictions['pred_labels_bertweet_and_hsol']]
bertweet_hsol_correct_predictions
```

```python
bertweet_hsol_correct_predictions.to_csv('bertweet_hsol_correct_predictions.csv')
```

```python
bertweet_hsol_incorrect_predictions = test_results_final[test_results_final['gt_labels_bertweet_hsol'] != test_results_final['pred_labels_bertweet_and_hsol']]
bertweet_correct_hsol_incorrect_predictions = bertweet_hsol_incorrect_predictions[bertweet_hsol_incorrect_predictions['gt_labels_bertweet'] == bertweet_hsol_incorrect_predictions['pred_labels_bertweet']]
```
